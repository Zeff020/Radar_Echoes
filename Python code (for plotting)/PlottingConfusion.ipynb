```python
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt
```

```python
array = [[1.00,0.00],
        [0.73,0.27]]      
df_cm = pd.DataFrame(array, range(2),
                  range(2))
df_cm.columns = ['Meteors', 'Others']
df_cm.index = ['Meteors', 'Others']
#plt.figure(figsize = (10,7))
sn.set(font_scale=1.4)#for label size
A = sn.heatmap(df_cm, cmap="Blues", annot=True,annot_kws={"size": 20})
A.set_ylabel('True')
A.set_xlabel('Predicted')# font size
```

```python
fig = A.get_figure()
fig.savefig("output.png")
```
